## Modelo Conceitual

Uma empresa de distribuição de derivados de petróleo está construindo um esquema de transporte com vários terminais de distribuição espalhados em algumas cidades (A, B, C, D), cuja demanda de produto é conhecida.

### Demanda Mensal

| Mercado | Demanda Mensal (t/mês) |
| ------- | ---------------------- |
| A       | 150.000                |
| B       | 100.000                |
| C       | 200.000                |
| D       | 210.000                |

No atendimento desta demanda, a empresa pretende utilizar transporte marítimo e terrestre. Para tanto, é necessário instalar terminais marítimos ao longo da costa. Os portos escolhidos para instalação de possíveis terminais localizam-se em **P1**, **P2**, **P3** e **P4**.

Os **tipos de terminais** que poderão ser instalados nesses portos são:

| Tipo    | Capacidade (t/mês) | Custo Fixo de Operação (R$ milhões/mês) | Custo Variável (R$/t) |
| ------- | ------------------ | --------------------------------------- | --------------------- |
| **I**   | 100000             | 500000                                  | 6,5                   |
| **II**  | 200000             | 800000                                  | 5,3                   |
| **III** | 300000             | 1000000                                 | 4,5                   |

O combustível a ser distribuído na região poderá vir de duas refinarias distintas. A refinaria 1 tem uma capacidade de produzir adicionalmente 600.000 ton / mes, a um custo unitário de R$ 350,00 / ton. A segunda refinaria tem uma capacidade ociosa de 550.000 ton / mes, e com um custo unitário de R$ 370,00 / ton. O custo unitário de transporte é de R$ 5,60 / ton.km para o modo terrestre (entre terminais e mercados) e R$ 0,82 / ton.km para o modo marítimo (entre refinarias e terminais). As distâncias de transporte estão apresentadas no quadro abaixo.

**Distância de Transporte (km)**

| De     | Merc. A | Merc. B | Merc. C | Merc. D | Ref. 1 | Ref. 2 |
| ------ | ------- | ------- | ------- | ------- | ------ | ------ |
| **P1** | 150     | 140     | 160     | 120     | 800    | 1400   |
| **P2** | 200     | 130     | 140     | 120     | 900    | 1300   |
| **P3** | 180     | 170     | 130     | 160     | 1200   | 1000   |
| **P4** | 120     | 130     | 180     | 120     | 1300   | 900    |

Formule um modelo para determinação do esquema ótimo de distribuição do combustível e de implantação dos terminais. Resolva o modelo formulado e analise os resultados.


### Modelo Matemático

#### Parâmetros

##### Distâncias e Custos

$
\begin{aligned}
D*{ji}^{mar} &\quad \text{: distância (km) do porto } j \text{ à refinaria } i, \\
D*{jk}^{terr} &\quad \text{: distância (km) do porto } j \text{ ao mercado } k, \\
c*{mar} &= 0.82 \\
c*{terr} &= 5.60.
\end{aligned}
$

##### Custos de Produção nas Refinarias

$
\begin{aligned}
p_1 &= 350 \quad \text{(R\$/ton na refinaria 1)}, \\
p_2 &= 370 \quad \text{(R\$/ton na refinaria 2)}.
\end{aligned}
$

##### Capacidades e Custos dos Terminais

$
\begin{array}{c|c|c|c|c}
\text{Tipo } m & \text{Capacidade (t/mês)} & F_m \, (\text{R\$}) & c_m \, (\text{R\$/t}) & I_m \, (\text{R\$}) \\ \hline
\text{I} & 100\,000 & 500\,000 & 6.5 & \text{(a definir)} \\
\text{II} & 200\,000 & 800\,000 & 5.3 & \text{(a definir)} \\
\text{III} & 300\,000 & 1\,000\,000 & 4.5 & \text{(a definir)}
\end{array}
$

##### Capacidade de Produção das Refinarias

$
\begin{aligned}
\text{Cap}\_1 &= 600\,000 \quad \text{(ton/mês para a refinaria 1)}, \\
\text{Cap}\_2 &= 550\,000 \quad \text{(ton/mês para a refinaria 2)}.
\end{aligned}
$

##### Demanda dos Mercados

$
\begin{array}{c|c}
\text{Mercado } k & \text{Demanda (t/mês)} \\ \hline
A & 150\,000 \\
B & 100\,000 \\
C & 200\,000 \\
D & 210\,000
\end{array}
$

#### Indices

$
\begin{aligned}
i &= \text{refinarias (1, 2)} \\
j &=\text{portos (1, 2, 3, 4)} \\
k &= \text{mercados (A, B, C, D)} \\
m &= \text{tipos de terminais (I, II, III)}
\end{aligned}
$

#### Variáveis de decisão

$
\begin{aligned}
x_{ijm} &= \text{quantidade de combustível transportada da refinaria } i \text{ para o porto } j \text{ (ton)}  \text{ do tipo } m  \\
y_{jmk} &= \text{quantidade de combustível transportada do porto } k \text{ do tipo } m \text{ para o mercado } k  \\
z_{jm} &= \text{Igual a 1 se for instalado, no porto } j \text{, um terminal do tipo } m \text{; 0 caso contrário}
\end{aligned}
$

#### Função Objetivo

$
\begin{aligned}
\text{min } Z =\; & \sum*{i}\sum*{j}\sum*{m} \Big( c*{mar} \cdot D*{ji}^{mar} \cdot x*{ijm} \Big) + \sum*{j}\sum*{m}\sum*{k} \Big( c*{terr} \cdot D*{jk}^{terr} \cdot y*{jmk} \Big) \\ & + \sum*{j}\sum*{m} \Big( F*m \cdot z*{jm} + c*m \cdot \sum*{i} x\_{ijm} \Big) + \sum_{i}\Big( p_i \cdot \sum_{j}\sum_{m} x_{ijm} \Big)
\end{aligned}
$

#### Restrições

##### 1. Capacidade de Produção nas Refinarias

$
\begin{aligned}
\sum*{j}\sum*{m} x_ijm \leq \text{Cap}\_i, \quad \forall i \in \{1,2\}
\end{aligned}
$

##### 2. Satisfação da Demanda dos Mercados

$
\begin{aligned}
\sum*{j}\sum*{m} y\_{jmk} = \text{Dem}\_k, \quad \forall k \in \{A,B,C,D\}
\end{aligned}
$

##### 3. Equilíbrio de Fluxo nos Terminais

$
\begin{aligned}
\sum*{i} x*{ijm} = \sum*{k} y*{jmk}, \quad \forall j \in \{1,2,3,4\},\; \forall m \in \{\text{I}, \text{II}, \text{III}\}
\end{aligned}
$

##### 4. Capacidade dos Terminais

$
\begin{aligned}
\sum*{i} x*{ijm} \leq \text{Cap}_m \cdot z_{jm}, \quad \forall j,\; \forall m
\end{aligned}
$

##### 5. Instalação Única por Porto

$
\begin{aligned}
\sum*{m} z*{jm} \leq 1, \quad \forall j \in \{1,2,3,4\}
\end{aligned}
$

$
\begin{aligned}
x_{ijm} \ge 0 \quad \forall i, \, \forall j, \, \forall m \\
y_{jmk} \ge 0 \quad \forall j, \, \forall m, \, \forall k 
\end{aligned}
$

$
\begin{aligned}
z_{jm} \in \{0, 1\} \quad \forall j, \, \forall m
\end{aligned}
$

```python
import pyomo.environ as pyo
import time
```

```python
# Criação do modelo
model = pyo.ConcreteModel("Distribuicao_Combustivel")

# ------------------------------
# Indices e Conjuntos
# ------------------------------
# Refinarias:
model.I = pyo.Set(initialize=[1, 2])
# Portos:
model.J = pyo.Set(initialize=[1, 2, 3, 4])
# Mercados:
model.K = pyo.Set(initialize=["A", "B", "C", "D"])
# Tipos de terminais:
model.M = pyo.Set(initialize=["I", "II", "III"])

# ------------------------------
# Parâmetros
# ------------------------------
# Custos de transporte
model.c_mar = pyo.Param(initialize=0.82)  # R$/ton.km (transporte marítimo)
model.c_terr = pyo.Param(initialize=5.60)  # R$/ton.km (transporte terrestre)

# Distâncias
# D_mar: distância do porto j à refinaria i
D_mar_data = {
    (1, 1): 800,
    (1, 2): 1400,
    (2, 1): 900,
    (2, 2): 1300,
    (3, 1): 1200,
    (3, 2): 1000,
    (4, 1): 1300,
    (4, 2): 900,
}
model.D_mar = pyo.Param(model.J, model.I, initialize=D_mar_data)

# D_terr: distância do porto j ao mercado k
D_terr_data = {
    (1, "A"): 150,
    (1, "B"): 140,
    (1, "C"): 160,
    (1, "D"): 120,
    (2, "A"): 200,
    (2, "B"): 130,
    (2, "C"): 140,
    (2, "D"): 120,
    (3, "A"): 180,
    (3, "B"): 170,
    (3, "C"): 130,
    (3, "D"): 160,
    (4, "A"): 120,
    (4, "B"): 130,
    (4, "C"): 180,
    (4, "D"): 120,
}
model.D_terr = pyo.Param(model.J, model.K, initialize=D_terr_data)

# Custos de produção nas refinarias
p_data = {1: 350, 2: 370}  # R$/ton
model.p = pyo.Param(model.I, initialize=p_data)

# Capacidades de produção das refinarias (ton/mês)
Cap_ref_data = {1: 600000, 2: 550000}
model.Cap_ref = pyo.Param(model.I, initialize=Cap_ref_data)

# Demanda dos mercados (ton/mês)
Dem_data = {"A": 150000, "B": 200000, "C": 120000, "D": 240000}
model.Dem = pyo.Param(model.K, initialize=Dem_data)

# Parâmetros dos terminais:
# Capacidades dos terminais (ton/mês)
Cap_m_data = {"I": 100000, "II": 200000, "III": 300000}
model.Cap_m = pyo.Param(model.M, initialize=Cap_m_data)
# Custos fixos de instalação dos terminais (R$)
F_m_data = {"I": 500000, "II": 800000, "III": 1000000}
model.F_m = pyo.Param(model.M, initialize=F_m_data)
# Custo variável de operação dos terminais (R$/ton)
c_m_data = {"I": 6.5, "II": 5.3, "III": 4.5}
model.c_m = pyo.Param(model.M, initialize=c_m_data)

# ------------------------------
# Variáveis de Decisão
# ------------------------------
# x_{ijm} >= 0: toneladas de combustível da refinaria i para o porto j com terminal do tipo m
model.x = pyo.Var(model.I, model.J, model.M, domain=pyo.NonNegativeReals)

# y_{jmk} >= 0: toneladas de combustível do porto j com terminal do tipo m para o mercado k
model.y = pyo.Var(model.J, model.M, model.K, domain=pyo.NonNegativeReals)

# z_{jm} ∈ {0,1}: 1 se instalar o terminal do tipo m no porto j, 0 caso contrário
model.z = pyo.Var(model.J, model.M, domain=pyo.Binary)


# ------------------------------
# Função Objetivo
# ------------------------------
def objective_rule(model):
    # Transporte marítimo: refinarias -> portos
    transport_mar = sum(
        model.c_mar * model.D_mar[j, i] * model.x[i, j, m]
        for i in model.I
        for j in model.J
        for m in model.M
    )

    # Transporte terrestre: portos -> mercados
    transport_terr = sum(
        model.c_terr * model.D_terr[j, k] * model.y[j, m, k]
        for j in model.J
        for m in model.M
        for k in model.K
    )

    # Custo dos terminais: custo fixo + custo variável (operacional)
    terminais = sum(
        model.F_m[m] * model.z[j, m]
        + model.c_m[m] * sum(model.x[i, j, m] for i in model.I)
        for j in model.J
        for m in model.M
    )

    # Custo de produção nas refinarias
    prod = sum(
        model.p[i] * sum(model.x[i, j, m] for j in model.J for m in model.M)
        for i in model.I
    )

    return transport_mar + transport_terr + terminais + prod


model.obj = pyo.Objective(rule=objective_rule, sense=pyo.minimize)

# ------------------------------
# Restrições
# ------------------------------


# 1. Capacidade de Produção nas Refinarias
def cap_ref_rule(model, i):
    return sum(model.x[i, j, m] for j in model.J for m in model.M) <= model.Cap_ref[i]


model.CapRef = pyo.Constraint(model.I, rule=cap_ref_rule)


# 2. Satisfação da Demanda dos Mercados
def demanda_rule(model, k):
    return sum(model.y[j, m, k] for j in model.J for m in model.M) == model.Dem[k]


model.Demanda = pyo.Constraint(model.K, rule=demanda_rule)


# 3. Equilíbrio de Fluxo nos Terminais: o que chega no porto via refinarias deve ser distribuído aos mercados
def fluxo_rule(model, j, m):
    return sum(model.x[i, j, m] for i in model.I) == sum(
        model.y[j, m, k] for k in model.K
    )


model.Fluxo = pyo.Constraint(model.J, model.M, rule=fluxo_rule)


# 4. Capacidade dos Terminais: se instalado, a quantidade que chega não pode exceder a capacidade do terminal
def cap_terminal_rule(model, j, m):
    return sum(model.x[i, j, m] for i in model.I) <= model.Cap_m[m] * model.z[j, m]


model.CapTerminal = pyo.Constraint(model.J, model.M, rule=cap_terminal_rule)


# 5. Instalação Única por Porto: no máximo um terminal (de qualquer tipo) pode ser instalado em cada porto
def instalacao_unica_rule(model, j):
    return sum(model.z[j, m] for m in model.M) <= 1


model.InstalacaoUnica = pyo.Constraint(model.J, rule=instalacao_unica_rule)

# ------------------------------
# Escrita do Modelo em Arquivo
# ------------------------------
model.write("modelo_combustivel.lp", io_options={"symbolic_solver_labels": True})

# ------------------------------
# Resolução
# ------------------------------
solver = pyo.SolverFactory("appsi_highs")
results = solver.solve(model, tee=True)

# Exibindo resultados
print("\nStatus do solver:", results.solver.status)
print("Condição de terminação:", results.solver.termination_condition)
valor_obj = pyo.value(model.obj)
formatted = f"{valor_obj:,.2f}"
formatted = formatted.replace(",", "X").replace(".", ",").replace("X", ".")
print("Valor da Função Objetivo: R$", formatted)

print("\n=== Solução: Fluxos e Instalação dos Terminais ===")
print("Fluxo das refinarias para portos (x[i,j,m]):")
for i in model.I:
    for j in model.J:
        for m in model.M:
            val = pyo.value(model.x[i, j, m])
            if val is not None and val > 1e-6:
                print(f"  Refinaria {i} -> Porto {j} (Terminal {m}): {val:.2f} t")

print("\nFluxo dos portos para mercados (y[j,m,k]):")
for j in model.J:
    for m in model.M:
        for k in model.K:
            val = pyo.value(model.y[j, m, k])
            if val is not None and val > 1e-6:
                print(f"  Porto {j} (Terminal {m}) -> Mercado {k}: {val:.2f} t")

print("\nInstalação dos terminais (z[j,m]):")
for j in model.J:
    for m in model.M:
        val = pyo.value(model.z[j, m])
        print(f"  Porto {j}, Terminal {m}: {int(round(val))}")
```

```python
# Testar diferentes solvers e medir o tempo
solvers = ["appsi_highs", "gurobi", "glpk"]
results = {}

for solver_name in solvers:
    if pyo.SolverFactory(solver_name).available():
        print(f"=== Resolvendo com {solver_name} ===")

        # Clonar o modelo para evitar que um solver afete o outro
        model_copy = model.clone()

        solver = pyo.SolverFactory(solver_name)
        start_time = time.time()
        result = solver.solve(model_copy, tee=False)
        end_time = time.time()

        # Guardar os resultados
        results[solver_name] = {
            "tempo": end_time - start_time,
            "valor_objetivo": pyo.value(model_copy.obj),
        }
    else:
        print(f"Solver {solver_name} não está disponível.\n")

# Comparação dos tempos de execução
print("\n=== Comparação de Tempo de Execução ===")
for solver_name, data in results.items():
    print(f"{solver_name}: {data['tempo']:.6f} segundos")

# Comparação dos resultados
print("\n=== Comparação dos Resultados ===")
for solver_name, data in results.items():
    print(f"Solver: {solver_name}")
    print(f"Valor Objetivo: {data['valor_objetivo']:.6f}")
```
